```python
def generate_resume(jd_name, jd, tag):
    resume_prompt = (
        f"Create a professional {tag} resume for the following candidate:\n\n"
        f"Job Role: {jd}\n"
        f"take your examples for the names for the resume"
        f"Make it professional, concise, and tailored to the role of {jd_name}.\n "
        f"Include an objective, experience summary, and highlight the core skills effectively."
    )

    # Call the AI API to generate the resume
    api_response = api_client.chat.completions.create(
        model="meta-llama/Llama-Vision-Free",
        messages=[{"role": "user", "content": resume_prompt}]
    )

    # Extract and return the AI-generated resume
    return api_response.choices[0].message.content.strip()
```

```python
def generate_q_and_a(jd_name, jd, resume, tag):
    resume_prompt = (
        f"Create an {tag} Interview between candiate and HR FOR\n\n"
        f"take your examples for the names for the interview"
        f"Job Role: {jd}\n"
        f"to the role of {jd_name}.\n "
    )

    # Call the AI API to generate the resume
    api_response = api_client.chat.completions.create(
        model="meta-llama/Llama-Vision-Free",
        messages=[{"role": "user", "content": resume_prompt}]
    )

    # Extract and return the AI-generated resume
    return api_response.choices[0].message.content.strip()
```

```python
def create_excel_file(data, output_file="output.xlsx"):
    # Convert data to DataFrame
    df = pd.DataFrame(data)
    # Save to Excel
    df.to_excel(output_file, index=False)
    print(f"Data saved to {output_file}")
```
